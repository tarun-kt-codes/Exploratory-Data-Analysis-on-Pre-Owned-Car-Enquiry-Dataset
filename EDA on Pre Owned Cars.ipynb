**Importing all the required libraries**

```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

**Creating the dataframe by reading the CSV file**

```python
enq=pd.read_csv('jan enq.csv')
```

**Dropping Unecessary Features**

```python
enq.drop(["CUST_MOBILE","MSPIN","EMAIL_ID","ENQ_NUM","DEALER_NAME","DEALER CODE-FOR CODE","CUST_NAME","DLR_CITY","TEST_DRIVE_NO","DSE_NAME","DSE_DESIGNATION","DEALER CODE-FOR CODE"],axis=1,inplace=True)
```

**Visualising the spread of various budgets**

```python
sns.displot(enq['BUDGET'],bins=30)
```

**Visualising the spread of age of old car**

```python
sns.displot(enq['AGE_OF_OLD_CAR'])
```

**Visualising the relationship between the age of the car and budget using a hexplot**

```python
sns.jointplot(x='BUDGET',y='AGE_OF_OLD_CAR',data=enq,kind='hex')
```

**Visualising the count of various enquiry modes**

```python
sns.histplot(enq["ENQ_MODE"])
```

**Visualising the count of all types of enquiry statuses**

```python
sns.countplot(x="ENQ_STATUS",data=enq)
```

**Visualising the relationship between car models and budget through catplot**

```python
sns.catplot(x="MODEL",y="BUDGET",data=enq,aspect=3)
```

**Visualising the count of all types of enquiry source**

```python
sns.countplot(x="ENQ_SOURCE",data=enq)
```

**Bar Plot between the customer budget and their gender**

```python
sns.barplot(x='CUST_GENDER',y='BUDGET',data=enq)
```

**Bar Plot between Enquiry Mode and Budget**

```python
sns.barplot(x='ENQ_MODE',y='BUDGET',data=enq,estimator=np.std)
```

**Bar Plot between Car Type and Budget**

```python
sns.barplot(x='CAR_TYPE',y='BUDGET',data=enq,estimator=np.std)
```

**Visualising the budget of customers based on their gender across all days of the week through Box Plot** 

```python
sns.boxplot(x="ENQ_DAY", y="BUDGET", data=enq,palette='seismic',hue='CUST_GENDER')
```

**Strip Pot between enquiry mode and budget**

```python
sns.stripplot(x="ENQ_MODE",y="BUDGET", data=enq,jitter=True,hue='CUST_GENDER',palette='seismic',dodge=True)
```

**Visualising the budget  of all types of cars enquired by the customers based on gender through Box Plot**

```python
sns.boxplot(x="CAR_TYPE", y="BUDGET", data=enq,palette='seismic',hue='CUST_GENDER')
```

**Visualising the budget enquiries made through all enquiry mode based on enquiry status using Box Plot**

```python
sns.boxplot(x="ENQ_MODE", y="BUDGET", data=enq,palette='seismic',hue='ENQ_STATUS')
```

**Visualising the enquiries made on all days of the week on all car types using Heatmap**

```python
pvenq = enq.pivot_table(values='BUDGET',index='ENQ_DAY',columns='CAR_TYPE')
sns.heatmap(pvenq,cmap='coolwarm',linecolor='white',linewidths=2)
```

**Linear Model Plot between age of old car and budget based on gender**

```python
sns.lmplot(x='BUDGET',y='AGE_OF_OLD_CAR',data=enq,hue='CUST_GENDER',palette='seismic')
```

**Linear Model Plots between budget and age of old car based on customer gender and car type**

```python
sns.lmplot(x='BUDGET',y='AGE_OF_OLD_CAR',data=enq,col='CUST_GENDER',row='CAR_TYPE')
```

**Linear Model Plots between budget and age of old car based on customer gender and enquiry day**

```python
sns.lmplot(x='BUDGET',y='AGE_OF_OLD_CAR',data=enq,col='CUST_GENDER',row='ENQ_DAY')
```
